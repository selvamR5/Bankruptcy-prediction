```python


from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
polish_companies_bankruptcy = fetch_ucirepo(id=365) 
  

  
```

```python
# libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
type(polish_companies_bankruptcy)
```

```python
polish_companies_bankruptcy.variables
```

```python
# Assign feature in DF
df = polish_companies_bankruptcy.data.features
print(type(df))
print(df.shape)
df.head()
```

```python
# Assign target in y
y = polish_companies_bankruptcy.data.targets
print(type(y))
print(y.shape)
y.head()
```

```python
df = pd.concat([df,y],axis = 1)
```

```python
print(df.shape)
df.head()
```

```python
df.info()
```

```python
# finding null values
print(df.shape)
df.isna().sum()
```

```python
# Removing na values
df = df[df["A27"] >= -259000.0]
df.shape

```

```python
# class balance 
print(df["class"].value_counts(normalize = True))
df["class"].value_counts(normalize = True).plot(kind = "bar")
plt.xlabel("Bankruptcy")
plt.ylabel("frequency")
plt.title("class balance")

# positive class - bankruptcy = true
# negative class - bankruptcy = no
```

```python
# side by side boxplot for bankruptcy and A27(profit by expense)
sns.boxplot(x = "class", y = "A27", data =  df)
plt.xlabel("bankruptcy")
plt.ylabel("profit/expense")
```

```python
# A27 describe (mean, std)
df["A27"].describe().apply("{0:,.0f}".format)
```
